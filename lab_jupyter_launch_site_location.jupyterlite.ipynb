```python
# Add circles and text label icons for each launch site
for index, site in launch_sites_df.iterrows():
    coordinates = [site['Lat'], site['Long']]  # Corrected coordinates assignment
    site_name = site['Launch Site']
    
    # Create a circle at the launch site's coordinates
    circle = folium.Circle(coordinates, radius=1000, color='blue', fill=True)
    circle.add_child(folium.Popup(site_name))  # Add popup label to the circle
    
    # Create a text label icon
    text_icon = folium.map.Marker(
        location=coordinates,
        icon=folium.DivIcon(
            icon_size=(30, 30),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12pt; color: blue;"><b>{site_name}</b></div>'
        )
    )
    
    # Add the circle and text label icon to the map
    site_map.add_child(circle)
    site_map.add_child(text_icon)

site_map
```

```python
from folium.plugins import MarkerCluster

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Function to get marker color based on 'class' column
def get_marker_color(row):
    return 'green' if row['class'] == 1 else 'red'

# Create and add markers to the MarkerCluster
for index, record in spacex_df.iterrows():
    coordinates = [record['Lat'], record['Long']]  # Assuming 'Lat' and 'Long' are columns in spacex_df
    marker_color = get_marker_color(record)
    
    # Create a marker with a specific color based on launch outcome
    marker = folium.Marker(location=coordinates, icon=folium.Icon(color='white', icon_color=marker_color))
    
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(marker)

site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
